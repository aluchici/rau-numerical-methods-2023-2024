```python
import numpy as np 

class Particle:
    def __init__(self, x, y):
        self.x = x
        self.y = y 
        self.vx = None 
        self.vy = None 

    def move(self, dt=0.1):
        self.vx = np.random.random()
        self.vy = np.random.random()

        self.x = self.x + self.vx * dt 
        self.y = self.y + self.vy * dt

    def is_colliding(self, particle):
        if self.x == particle.x and self.y == particle.y:
            return True 
        return False 
    
class Particle3D(Particle):
    def __init__(self, x, y, z):
        super().__init__(x, y)
        self.z = z

        self.vz = None

    def is_colliding(self, particle):
        if self.x == particle.x and self.y == particle.y and self.z == particle.z:
            return True
        else:
            return False
```

```python
particle1 = Particle(10, 10)
particle2 = Particle(20, 20)

collisions = []
trajectory = []
for i in range(100):
    trajectory.append(((particle1.y, particle1.x), (particle2.y, particle2.x)))
    particle1.move()
    particle2.move()
    collision = particle1.is_colliding(particle2)
    collisions.append(collision)

print(f"Number of collisions = {np.sum(collisions)}")
```

```python
particle3d_1 = Particle3D(1, 2, 3)
particle3d_2 = Particle3D(1, 2, 3)

particle3d_1.is_colliding(particle3d_2)
```

```python
import matplotlib.pyplot as plt 

_ = plt.figure(figsize=(10, 5))

x1 = [t[0][1] for t in trajectory]
y1 = [t[0][0] for t in trajectory]

x2 = [t[1][1] for t in trajectory]
y2 = [t[1][0] for t in trajectory]

plt.plot(x1, y1, c="r", label="Particle 1")
plt.scatter(x1, y1, s=5, c="r")

plt.plot(x2, y2, c="b", label="Particle 2")
plt.scatter(x2, y2, s=5, c="b")
plt.legend()
plt.show()
```
