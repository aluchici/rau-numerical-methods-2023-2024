Dynamics of an ideal Savings Account.

System dynamics:

* dA/dt = rA + D - W 
* A = account balance
* dA/dt = monthly rate of change of the account balance 
* r = interest rate (e.g. 0.05, constant)
* D = deposit (monthly, constant)
* W = withdrawal (monthly, constant)
* P = initial account balance (month 0)

First approximation: 

A_{t+1} = A_t + dt * (rA + D - W)

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Setup the simulation with constant D, W values
r = 0.03 
D = 100
W = 50
P = 100
```

```python
# Implement the solution 
n_iters = 36 # number of months 
dt = 1 # 1 month 
A = [P]
for i in range(1, n_iters):
    Ai = A[i-1] + dt * (r*A[i-1] + D - W)
    A.append(Ai)
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(range(n_iters), A, c="magenta", linestyle="-.")
plt.scatter(range(n_iters), A, s=50)
plt.xlabel("Month #")
plt.ylabel("Account value (EUR)")
plt.title("Monthly account value")
plt.show()
```

```python
# Setup simulation with varying D, W
dt = 1
n_iters = 30 * 12
P = 100
r = 0.03
D = np.random.random(n_iters) * 100 
W = np.random.random(n_iters) * 100 * 0
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(range(n_iters), D, c="magenta", linestyle="-.", label="Deposits")
plt.scatter(range(n_iters), D, s=50, c="magenta")
plt.plot(range(n_iters), W, c="green", linestyle="-.", label="Withdrawls")
plt.scatter(range(n_iters), W, s=50, c="green")
plt.xlabel("Month #")
plt.ylabel("Amount (EUR)")
plt.title("Deposits vs Withdrawls")
plt.show()
```

```python
A = [P]
for i in range(1, n_iters):
    Ai = A[i-1] + dt * (r*A[i-1] + D[i-1] - W[i-1])
    A.append(Ai)
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(range(n_iters), A, c="magenta", linestyle="-.")
plt.scatter(range(n_iters), A, s=50, c="magenta")
plt.xlabel("Month #")
plt.ylabel("Account value (EUR)")
plt.title("Monthly account value")
plt.show()
```

```python
# Setup simulation with varying D, W
dt = 1
n_iters = 60
P = 100
r = 0.05
D = np.random.random(n_iters)
W = np.random.random(n_iters)
```

```python
A = [P]
for i in range(1, n_iters):
    Ai = A[i-1] + dt * (r*A[i-1] + D[i-1] * A[i-1] - W[i-1] * A[i-1])
    A.append(Ai)
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(range(n_iters), A, c="magenta", linestyle="-.")
plt.scatter(range(n_iters), A, s=50, c="magenta")
plt.xlabel("Month #")
plt.ylabel("Account value (EUR)")
plt.title("Monthly account value")
plt.show()
```
