```python
import numpy as np
```

```python
v = np.array([3, 3])
w = np.array([-3, 3])
u = v + w
t = v - w
v2 = 2 * v
```

```python
import matplotlib.pyplot as plt 

origin = np.array([0, 0])

_ = plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], scale=21)
plt.quiver(0, 0, w[0], w[1], scale=21, color='g')
plt.quiver(0, 0, u[0], u[1], scale=21, color='r')
plt.quiver(0, 0, t[0], t[1], scale=21, color='b')
plt.quiver(0, 0, v2[0], v2[1], scale=21, color='m')
# plt.quiver(origin, w)
plt.show()
```

```python
np.dot(v, w), np.dot(v, u)
```

```python
A = np.array([[1, 2],
              [3, -1]
              ])
v3 = A.dot(v)
v3
```

```python
v3 = A.dot(v3)
v3
```

```python
vs = []
for i in range(10):
    v3 = A.dot(v3)
    vs.append(v3)

_ = plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], scale=21)
plt.quiver(0, 0, v3[0], v3[1], scale=21, color='m')
for i in range(10): 
    plt.quiver(0, 0, vs[i][0], vs[i][1], scale=21, color='m')
plt.show()
```

```python
v = np.array([2, 1])
t = [-5, -1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1, 5]
vs = []
for ti in t:
    vs.append(ti * v)

for i in range(len(t)): 
    plt.quiver(0, 0, vs[i][0], vs[i][1], scale=21, color='m')
plt.show()
```

```python
# R = [[cos a, -sin a], [sin a, cos a]]
v = np.array([3, 2])
a = np.linspace(-np.pi, np.pi, 25)
vs = []
for i in range(a.shape[0]):
    A = np.array([[np.cos(a[i]), -np.sin(a[i])],
                  [np.sin(a[i]), np.cos(a[i])]])
    vi = A.dot(v)
    vs.append(vi)

for i in range(a.shape[0]): 
    plt.quiver(0, 0, vs[i][0], vs[i][1], scale=21, color='m')
plt.show()   
```
